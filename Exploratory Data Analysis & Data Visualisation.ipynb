# Data Visualisation

```python
def stacked_bars(dataframe, title_, size_ = (18, 10), rot_ = 0, legend_ = "upper right"):
    ax = dataframe.plot(kind = "bar", 
                        stacked=True, 
                        figsize=size_, 
                        rot=rot_, 
                        title=title_)
    # annotate bars
    annotate_stacked_bars(ax, textsize = 14)
    # rename legend
    plt.legend(["Retention","Churn"], loc = legend_)
    # labels
    plt.ylabel("Company base (%)")
    plt.show()

def annotate_stacked_bars(ax, pad = 0.99, colour="white", textsize =13):
    # iterate over the plotted bars
    for p in ax.patches:
        #calculate annotations
        value = str(round(p.get_height(),1))
        #if value is zero, do not annotate
        if value == "0.0":
            continue
        ax.annotate(value,
                    ( (p.get_x() + p.get_width()/2)*pad-0.05, (p.get_y() + p.get_height()/2)*pad ),
                   color = colour,
                   size = textsize)    
```

# Consumption

Let's see the distribution of the consumption in the last year and month. Since the consumption data is univariate, let's use histograms to visualize their distribution.

```python
consumption = client[['id', 'cons_12m', 
                      'cons_gas_12m', 'cons_last_month', 
                      'imp_cons', 'has_gas', 'churn']]
#id = client company identifier
#cons_12m = electricity consumption of the past 12 months
#cons_gas_12m = gas consumption of the past 12 months
#cons_last_month = electricity consumption of the last month
#imp_cons = current paid consumption
#has_gas = indicated if client is also a gas client
#churn = has the client churned over the next 3 months
consumption.head(3)
```

```python
def plot_distribution(dataframe, column, ax, bins_=50):
    
    # Create a temporal dataframe with the data to be plot
    temp = pd.DataFrame(
        {"Retention": dataframe[dataframe["churn"]==0][column],
         "Churn":dataframe[dataframe["churn"]==1][column]})
    
    # Plot the histogram
    temp[["Retention","Churn"]].plot(kind='hist', bins=bins_, ax=ax, stacked=True)
    
    # X-axis label
    ax.set_xlabel(column)
    # Change the x-axis to plain style
    ax.ticklabel_format(style='plain', axis='x')

#bins_=50 => your values range from 0 to 500,000; 
            #Each bin will span about 10,000 units (500000 / 50)
            #More bins = more detail, but may look noisy
            #Fewer bins = smoother, but may hide patterns
#Retention = customers who did not churn (churn == 0)
#Churn = customers who did churn (churn == 1)
#Stacked=True => lets you visualize overlap
        #It stacks the "Retention" and "Churn" bars on top of each other for each bin
#ax=ax => tells the .plot() function where to draw the histogram.
        #Draw this histogram in the first subplot (axs[0]).
```

```python
fig, axs = plt.subplots(nrows=4, figsize=(18, 25))

plot_distribution(consumption, 'cons_12m', axs[0])
plot_distribution(consumption[consumption['has_gas'] == 't'], 'cons_gas_12m', axs[1])
plot_distribution(consumption, 'cons_last_month', axs[2])
plot_distribution(consumption, 'imp_cons', axs[3])

plt.tight_layout()   # Avoid overlapping
plt.show()

#Creates 4 vertical subplots (shared axs array)
#figsize=(18, 25) â†’ large figure for clarity
```

### 1. Electricity Consumption Over 12 Months (cons_12m)

#### Key Observations:
Very right-skewed distribution â€” the vast majority of customers use under ~100,000 units.
A long tail reaches above 6 million, but those are extreme outliers.
Most churners (orange) are also clustered in the low consumption range.
The largest bin (leftmost) contains most of the data, with both churners and retainers.

#### Interpretation:
Churn is not significantly higher among high-consumption customers.
Electricity consumption alone may not be a strong predictor of churn unless combined with other variables (e.g., billing issues, seasonal behavior, or sudden changes).

### 2. Gas Consumption Over 12 Months (cons_gas_12m)

#### Key Observations:
Also right-skewed, but with fewer customers compared to cons_12m, since this only includes customers with has_gas = True.
Most clients have very low gas consumption.
Churn (orange) is concentrated in the lowest bin, similar to electricity.
Fewer extreme outliers, but still present (up to 4 million).

#### Interpretation:
Churners among gas users behave similarly to electricity users â€” mostly low consumers.
Having gas doesn't indicate higher or lower churn on its own.
Segmenting by industry or size could reveal more.

### 3. Electricity Consumption Last Month (cons_last_month)

#### Key Observations:
The pattern is very similar to the 12-month view, which makes sense since cons_last_month is part of cons_12m.
Very steep drop after the first bin.
Outliers exist but are sparse.
Churn again is concentrated in the low range.

#### Interpretation:
No significant "spike" or deviation in churners' last-month consumption.
Monthly consumption alone doesnâ€™t seem to cause churn â€” maybe sudden change or discrepancy between expected vs. billed usage would matter more.

### 4. Billed Consumption (imp_cons)

#### Key Observations:
Still right-skewed, but with a much smaller range (~0 to 15,000).
The distribution is more compact and much denser in the 0â€“500 range.
Most churners are again located in the lowest bin, overlapping with retained clients.

#### Interpretation:
Most clients are billed very low values.
Low or no billed usage might correlate with service disconnection or inactivity, which could trigger churn.

This is the narrowest and cleanest distribution so far â€” could be valuable as a binary churn signal (e.g., if imp_cons == 0, higher risk?).

## Margins(Profits)

```python
margin = client[['id', 'margin_gross_pow_ele', 'margin_net_pow_ele', 'net_margin']]
#margin_gross_pow_ele = gross margin on power subscription
#margin_net_pow_ele = net margin on power subscription
#net_margin = total net margin
```

```python
fig, axs = plt.subplots(nrows=3, figsize=(18,20))
# Plot histogram
sns.boxplot(x = margin["margin_gross_pow_ele"], ax=axs[0])
sns.boxplot(x = margin["margin_net_pow_ele"],ax=axs[1])
sns.boxplot(x = margin["net_margin"], ax=axs[2])

plt.show()
```

### 1. margin_gross_pow_ele (Gross Margin on Power Subscription)
This is how much money the company earns from selling electricity to a customer, before subtracting any costs (like staff or service expenses).
Think of it as the initial profit from selling electricity.
The distribution is right-skewed (long tail on the right).
The box shows the interquartile range (IQR) tightly packed between approximately 10 and 50.
There are many outliers extending well beyond 50, with some reaching as high as 350+.
This suggests a majority of customers have gross margins under 50, but a few have extremely high margins.

### 2. margin_net_pow_ele (Net Margin on Power Subscription)
This is how much money the company really makes from electricity sales after costs (like operations or delivery fees) are subtracted.
Itâ€™s the actual profit from the power subscription.
The shape and characteristics are very similar to the gross margin:
Also right-skewed.
IQR is around 10 to 50.
Many outliers, possibly the same customers with high gross margins.
This consistency makes sense, as net margin is often derived from gross margin with deductions like operating costs.

### 3. net_margin (Total Net Margin)
This is the total real profit from the customer, including all services, not just electricity â€” it could include gas, service fees, or other subscriptions too. 
Much more extremely skewed to the right.
Most of the data lies between 0 and ~500, but there are some very extreme outliers, with values going above 25,000.
This suggests that while most customers generate a relatively modest total net margin, a few large customers are contributing disproportionately high margins (likely due to multiple services or very large contracts).
